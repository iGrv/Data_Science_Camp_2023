# Home task: task about lakes

- Generate 20 random variable from 1 to 100
- Draw the plot of the  sequence 
- Lets assume each point represent the height and so all plotting is 2d mountains. 
- Then consider the unlimited rain from above - cavities become lakes full of water. 
- Determine the deepest lake

### Expected result

You may present the result by  visualization or numbers that provide information about the boundaries of the lake and its deepest point 

```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.signal import argrelextrema
%matplotlib inline
```

```python
# Generate a numpy array of 20 random integer variables in interval [1, 100] (use 2023 as a seed)
np.random.seed(2023)
rand_array = np.random.randint(1, 101, size=20)

# Show the plot of generated array
plt.plot(rand_array);
```

```python
def lake_boundaries(mountains, approx_equal):
    """
    Determines boundaries of the deepest lake in `mountains` formed by rain from above

    :param mountains: array of numbers that can be represented in the plot as heights of mountains
    :type mountains: NumPy ndarray

    :param approx_equal: function with two parameters that determines whether these parameters
    are "approximately equal" as heights of mountains (note: function must use standard NumPy
    functions instead of their built-in analogs, e.g. min, max, sum, abs, any, all etc)
    :type approx_equal: function

    :return: three values: first value is an index of bottom of the deepest lake in `mountains` array,
    second value is index of left boundary of the deepest lake, and third value is index of right
    boundary of the deepest lake
    :rtype: tuple[int, int, int]
    """
    # Bottom of the deepest lake is located where index of minimum value in mountains array is
    bottom = mountains.argmin()

    # Possible boundaries of the deepest lake must be the peaks in mountains array

    # Indices of local maximums in mountains array represent the locations of mountains peaks
    peaks, = argrelextrema(mountains, np.greater_equal)  # Includes also indices of flat peaks

    # Consider when bottom of the deepest lake located at the beginning or at the end of mountains array
    if bottom in [0, (mountains.size - 1)]:
        
        # Find indices of high peaks (i.e. peaks that is "approximately equal" to the highest peak)
        high_peaks = approx_equal(mountains[peaks], mountains.max())

        # If bottom located at the beginning of mountains array, then
        # left boundary is bottom and right boundary is first left high peak
        if bottom == 0:
            left = bottom
            right = peaks[high_peaks][0]

        # If bottom located at the end of mountains array, then
        # left boundary is first right high peak and right boundary is bottom
        else:
            left = peaks[high_peaks][-1]
            right = bottom
    
    # Consider when bottom of the deepest lake located in the middle of mountains array
    else:

        # Get indices of peaks to the left and to the right of bottom
        left_peaks = peaks[peaks < bottom]
        right_peaks = peaks[peaks > bottom]

        # Find indices of high peaks ("approximately equal" to the highest peak)
        # to the left and to the right of bottom
        high_lefts = approx_equal(mountains[left_peaks], mountains.max())
        high_rights = approx_equal(mountains[right_peaks], mountains.max())

        # There must at least one those peak (e.g. highest peak) to the left or to the right of bottom

        # Consider when there are no high peaks to the left of bottom, but there are some to the right
        if not np.any(high_lefts):

            # Determine right boundary as first left peak of right high peaks
            right = right_peaks[high_rights][0]

            # Find indices of peaks to the left of bottom that are
            # "approximately equal" to the determined right boundary
            eq_lefts = approx_equal(mountains[left_peaks], mountains[right])

            # Determine left boundary as first right peak of "approximately equal" peaks to the left of bottom
            # if there are those, otherwise, determine left boundary as first element in array
            left = left_peaks[eq_lefts][-1] if np.any(eq_lefts) else 0

        # Consider when there are no high peaks to the right of bottom, but there are some to the left
        elif not np.any(high_rights):

            # Determine left boundary as first right peak of left high peaks
            left = left_peaks[high_lefts][-1]

            # Find indices of peaks to the right of bottom that are
            # "approximately equal" to the determined left boundary 
            eq_rights = approx_equal(mountains[right_peaks], mountains[left])

            # Determine right boundary as first left peak of "approximately equal" peaks to the right of bottom
            # if there are those, otherwise, determine right boundary as last element in array
            right = right_peaks[eq_rights][0] if np.any(eq_rights) else mountains.size - 1
        
        # Consider when there are high peaks to the left and to the right of bottom
        else:
            
            # Determine left boundary as first right peak of left high peaks
            left = left_peaks[high_lefts][-1]

            # Determine right boundary as first left peak of right high peaks
            right = right_peaks[high_rights][0]

    return bottom, left, right
```

```python
# Find bottom, left and right boundaries of the deepest lake
# For the "approximately equal" check use whether absolute difference between
# two values is less than 0.1 of range of values in generated array
bottom, l, r = lake_boundaries(rand_array, lambda x, y: np.abs(x - y) < 0.1 * np.ptp(rand_array))
print("Bottom of the deepest lake is at index", bottom)

# Show the plot of generated array and the deepest lake
lake = np.arange(l, r + 1)
plt.plot(rand_array)
plt.plot(lake, rand_array[lake], color="red");
```
