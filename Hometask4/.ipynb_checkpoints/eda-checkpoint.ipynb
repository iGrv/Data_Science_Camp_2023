# Home task: Exploratory Data Analysis (EDA) of car insurance dataset

__Dataset description:__ Car insurance claim. As you are working in the insurance company. Company wants to know the reason why claim was not made. Doing so would allow insurance company to improve there policy for giving loan to the customer. In this project you are dealing with various feature such as age, occupation etc. based on that let's get back to the final conculsion.

__Steps:__
- Data Cleaning
- Descriptive Statistics
- Data Visualization
- Conclusions

```python
import matplotlib.pyplot as plt
import seaborn as sns  # Version 0.13.0
import pandas as pd
plt.style.use("ggplot")

%matplotlib inline
FILE = "Car_Insurance_Claim.csv"
COLORS = plt.rcParams['axes.prop_cycle'].by_key()["color"]
```

```python
# Load dataset from CSV file
car_insurance = pd.read_csv(FILE, index_col=0, na_values=["none", ""])
car_insurance.info()
```

### Cleaning data

```python
# Convert POSTAL_CODE data as string type
car_insurance["POSTAL_CODE"] = car_insurance["POSTAL_CODE"].astype("object")

# Convert columns which contain categorical data to category type
category_cols = ["AGE", "GENDER", "RACE", "DRIVING_EXPERIENCE", "EDUCATION", "INCOME", "VEHICLE_YEAR", "VEHICLE_TYPE"]
for col in category_cols:
    car_insurance[col] = car_insurance[col].astype("category")

# Convert columns from numeric to category type
car_insurance["VEHICLE_OWNERSHIP"] = car_insurance["VEHICLE_OWNERSHIP"].map({1: "owner", 0: "not owner"}).astype("category")
car_insurance["MARRIED"] = car_insurance["MARRIED"].map({1: "married", 0: "unmarried"}).astype("category")
car_insurance["CHILDREN"] = car_insurance["CHILDREN"].map({1: "has children", 0: "has no children"}).astype("category")
car_insurance["OUTCOME"] = car_insurance["OUTCOME"].map({1: "insurance", 0: "no insurance"}).astype("category")

# Rename category names in AGE and DRIVING_EXPERIENCE columns
car_insurance["AGE"] = car_insurance["AGE"].apply(lambda x: f"{x} years")
car_insurance["DRIVING_EXPERIENCE"] = car_insurance["DRIVING_EXPERIENCE"].map({
    "0-9y": "0-9 years", "10-19y": "10-19 years", 
    "20-29y": "20-29 years", "30y+": "30+ years"
})

# Check dataset values
car_insurance.head()
```

So, the dataset has 19 columns:
 1. `ID` - entry's ID
 2. `AGE` - age of the customer (splitted into 4 categories: "16-25 years", "26-39 years", "40-64 years" and "65+ years")
 3. `GENDER` - gender of the customer (can be "male" or "female")
 4. `RACE` - race of the customer (can be "majority" or "minority")
 5. `DRIVING_EXPERIENCE` - driving experience of the customer (splitted into 4 categories: "0-9 years", "10-19 years", "20-29 years" and "30+ years")
 6. `EDUCATION` - education of the customer (can be "high school" or "university")
 7. `INCOME` - social status of the customer based on their income (splitted into categories "poverty", "working class", "middle class" and "upper class")
 8. `CREDIT_SCORE` - credit score of the customer (data from the bank)
 9. `VEHICLE_OWNERSHIP` - whether the customer is vehicle's owner (can be "owner" or "not owner")
 10. `VEHICLE_YEAR` - model year of the vehicle (can be "before 2015" or "after 2015")
 11. `MARRIED` - whether the customer is married (can be "married" or "unmarried")
 12. `CHILDREN` - whether the customer has children (can be "has children" or "has no children")
 13. `POSTAL_CODE` - postal code that idetifies the postal address of the customer
 14. `ANNUAL_MILEAGE` - average number of miles that vehicle has travelled in a year
 15. `VEHICLE_TYPE` - type of the vehicle (splitted into 2 categories: "sedan" and "sports car")
 16. `SPEEDING_VIOLATIONS` - number of speed violations commited by the customer
 17. `DUIS` - number of DUI (driving under the influence) incidents commited by the customer
 18. `PAST_ACCIDENTS` - number of past accidents the customer has
 19. `OUTCOME` - target value, whether insurance claim was made (can be "insurance" or "no insurance")

`ID` and `POSTAL_CODE` columns are not important for further data analys, thus 17 columns remain in the dataset.

##### Data distribution

We can use bar charts (for categorical data) and histograms (for numerical data) to visualize which possible values columns have.

```python
def bar(ax, data, column, order=None):
    """
    Shows vertical bar chart of dataset column that
    displays the data count in distinct categories.
    """
    sns.countplot(ax=ax, data=data, x=column, hue=column, order=order, legend=False)
    ax.set_ylabel("Count")
    for countainer in ax.containers:
        ax.bar_label(countainer)

def hist(ax, data, column, color=None, bins="auto", discrete=None):
    """
    Shows histogram of dataset column which contains numerical data.
    """
    sns.histplot(ax=ax, data=data, x=column, color=color, bins=bins, discrete=discrete)
    ax.set_ylabel("Count")
```

##### Data grouped by OUTCOME values

We can build grouped diagrams to display value distibutions of columns grouped by `OUTCOME`, i.e. in cases where insurance claim was made and where insurance claim was not made.

```python
def by_bar(ax, data, column, by, title, order=None):
    """
    Shows vertical count bar chart of dataset column
    grouped by values of other dataset column.
    """
    sns.countplot(ax=ax, data=data, x=column, hue=by, order=order)
    ax.set_title(title)
    ax.set_ylabel("Count")
    for countainer in ax.containers:
        ax.bar_label(countainer)

def by_hist(ax, data, column, by, title, bins="auto", discrete=None):
    """
    Shows dataset column histograms grouped
    by values of other dataset column.
    """
    sns.histplot(ax=ax, data=data, x=column, hue=by, bins=bins, discrete=discrete)
    ax.set_title(title)
    ax.set_ylabel("Count")
    
def by_box(data, column, by, fill=False):
    """
    Shows box plots of dataset columns grouped
    by values of other dataset column.
    """
    sns.boxplot(data=data, x=column, y=by, hue=by, fill=fill, legend=False, whis=(0, 100))

def by_title(text, by_1="insurance", by_2="no insurance"):
    return f"{text}: {by_1} vs {by_2}"
```

```python
# Build grouped plots for columns from first group (5 plots in total: 2 plots vertically and 3 plots horizontally)
fig, ax = plt.subplots(3, 2, figsize=(15, 20))
fig.suptitle(by_title("General data about customer"), y=0.93, fontsize="xx-large")
fig.delaxes(ax[2, 1])

by_bar(ax=ax[0, 0], data=car_insurance, column="AGE", by="OUTCOME", title=by_title("AGE"))
by_bar(ax=ax[0, 1], data=car_insurance, column="GENDER", order=["male", "female"], by="OUTCOME", title=by_title("GENDER"))
ax[0, 1].legend(title="OUTCOME", loc="upper left")
by_bar(ax=ax[1, 0], data=car_insurance, column="RACE", by="OUTCOME", title=by_title("RACE"))
by_bar(ax=ax[1, 1], data=car_insurance, column="MARRIED", by="OUTCOME", title=by_title("MARRIED"))
by_bar(ax=ax[2, 0], data=car_insurance, column="CHILDREN", by="OUTCOME", title=by_title("CHILDREN"))
```

We can see on "AGE: insurance vs no insurance" bar chart that insurance claim was made more likely to younger customers (16-25 years) than to older customers.

"GENDER: insurance vs no insurance" bar chart: the ratio of "insurance" to "no insurance" is slightly higher for male customers than for female customers.

"RACE: insurance vs no insurance" bar chart: the ratio of "insurance" to "no insurance" is approximately equal for customers of majority and minority race.

"MARRIED: insurance vs no insurance" bar chart: the ratio of "insurance" to "no insurance" is lower for married customers than for unmarried customers.

"CHILDREN: insurance vs no insurance" bar chart: the ratio of "insurance" to "no insurance" is mush more lower for customers with children than for for customers without children.

```python
# Build grouped plots for columns from second group (4 plots in total: 2 plots vertically and 2 plots horizontally)
fig, ax = plt.subplots(2, 2, figsize=(15, 15))
fig.suptitle(by_title("Educational and finance data about customer"), y=0.93, fontsize="xx-large")

by_bar(ax=ax[0, 0], data=car_insurance, column="EDUCATION", by="OUTCOME", title=by_title("EDUCATION"))
ax[0, 0].legend(title="OUTCOME", loc="upper left")
by_bar(ax=ax[0, 1], data=car_insurance, column="INCOME", order=["poverty", "working class", "middle class", "upper class"],
       by="OUTCOME", title=by_title("INCOME"))
by_hist(ax=ax[1, 0], data=car_insurance, column="CREDIT_SCORE", by="OUTCOME", title=by_title("CREDIT_SCORE"), bins=15)
by_bar(ax=ax[1, 1], data=car_insurance, column="VEHICLE_OWNERSHIP", by="OUTCOME", title=by_title("VEHICLE_OWNERSHIP"))
```

"EDUCATION: insurance vs no insurance" bar chart: the ratio of "insurance" to "no insurance" is slightly higher for customers who have only completed high school than for customers who have completed university.

"INCOME: insurance vs no insurance" bar chart: insurance claim was made more likely to customers from poverty that to customers from other social classes.

"CREDIT_SCORE: insurance vs no insurance" histogram: median of credit score values of customers to whom insurance claim was made is lower than median of credit score values of customers to whom no insurance claim was made. We can see this in the box plot below.

"VEHICLE_OWNERSHIP: insurance vs no insurance" bar chart: insurance claim was made more likely to customers who do not own the vehicle, than to customers who do own the vehicle.

```python
plt.figure(figsize=(5, 5))
plt.title(by_title("Value distribution of CREDIT_SCORE"))
by_box(data=car_insurance, column="CREDIT_SCORE", by="OUTCOME", fill=False);
```

```python
# Build grouped plots for columns from third group (4 plots in total: 2 plots vertically and 2 plots horizontally)
fig, ax = plt.subplots(2, 2, figsize=(15, 15))
fig.suptitle(by_title("Data about customer's past driving experience"), y=0.93, fontsize="xx-large")

by_bar(ax=ax[0, 0], data=car_insurance, column="DRIVING_EXPERIENCE", by="OUTCOME", title=by_title("DRIVING_EXPERIENCE"))
by_hist(ax=ax[0, 1], data=car_insurance, column="SPEEDING_VIOLATIONS", by="OUTCOME",
        title=by_title("SPEEDING_VIOLATIONS"), discrete=True)
by_hist(ax=ax[1, 0], data=car_insurance, column="DUIS", by="OUTCOME", title=by_title("DUIS"), discrete=True)
by_hist(ax=ax[1, 1], data=car_insurance, column="PAST_ACCIDENTS", by="OUTCOME", title=by_title("PAST_ACCIDENTS"), discrete=True)
```

##### Relationships between data

We can find relationships between dataset columns by calculating their correlation coefficient. To do this, we need to convert all column data to a numeric type. We also need to eliminate from calculations columns `ID` and `POSTAL_CODE`.

```python
# Delete columns ID and POSTAL_CODE from dataset
converted = car_insurance.drop(["ID", "POSTAL_CODE"], axis="columns")

# Convert categorical data to numeric type
to_numeric = {
    "AGE": {"16-25 years": 0, "26-39 years": 1, "40-64 years": 2, "65+ years": 3},
    "GENDER": {"male": 0, "female": 1},
    "RACE": {"majority": 0, "minority": 1},
    "DRIVING_EXPERIENCE": {"0-9 years": 0, "10-19 years": 1, "20-29 years": 2, "30+ years": 3},
    "EDUCATION": {"high school": 0, "university": 1},
    "INCOME": {"poverty": 0, "working class": 1, "middle class": 2, "upper class": 3},
    "VEHICLE_OWNERSHIP": {"not owner": 0, "owner": 1},
    "VEHICLE_YEAR": {"before 2015": 0, "after 2015": 1},
    "MARRIED": {"unmarried": 0, "married": 1},
    "CHILDREN": {"has no children": 0, "has children": 1},
    "VEHICLE_TYPE": {"sedan": 0, "sports car": 1},
    "OUTCOME": {"no insurance": 0, "insurance": 1}
}
for column, mapping in to_numeric.items():
    converted[column] = converted[column].map(mapping).astype(float)

# Check dataset values
converted.head()
```

Visualization of the columns correlation coefficients can be carried out by building a heat map.

```python
fig = plt.figure(figsize=(15, 12))
fig.suptitle("Correlation between dataset columns", y=0.92, fontsize="xx-large")
sns.heatmap(data=converted.corr(), annot=True, fmt=".2f", linewidths=0.1, cmap="seismic", vmin=-1, vmax=1);
```

 - For `CREDIT_SCORE` there is negative correlation (-0.33): insurance claim can be made more likely to customers with lower credit score
 - For `VEHICLE_OWNERSHIP` there is negative correlation (-0.38): insurance claim can be made more likely to customers who do not own the vehicle
 - For `VEHICLE_YEAR` there is negative correlation (-0.29): insurance claim can be made more likely to customers with vehicles produced before 2015
 - For `MARRIED` there is negative correlation (-0.26): insurance claim can be made less likely to married customers
 - For `CHILDREN` there is negative correlation (-0.23): insurance claim can be made less likely to customers with children
 - For `ANNUAL_MILEAGE` there is low positive correlation (0.19): insurance claim can be made more likely to customers with vehicles which have slightly higher annual mileage
 - For `VEHICLE_TYPE` correlation is close to zero (0.01) [note: sedan - 0, sports car - 1]: vehicle type does not have significant effect on the `OUTCOME` value

 - For `PAST_ACCIDENTS` the is negative correlation (-0.33): insurance claim can be made more likely to customers have commited less past accidents or have not commited any past accidents

Therefore, the columns which have the highest effect on the `OUTCOME` value are `AGE`, `DRIVING_EXPERIENCE`, `INCOME` and `VEHICLE_OWNERSHIP`.
