## Home task: Principal Component Analysis (PCA)
</font>

Apply PCA to breast cancer dataset and visualize 2 principle components.

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler

# Load features and labels from the breast cancer dataset
X, y = load_breast_cancer(return_X_y=True)

# Normalize the data using standard score
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.decomposition import PCA

# Perform PCA to get 2 principal components from the breast cancer dataset
pca = PCA(n_components=2)
Z = pca.fit_transform(X_scaled)

print('X_scaled.shape =', X_scaled.shape)
print('Z.shape =', Z.shape)
```

```python
# Visualize the principal components
plt.figure()
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.scatter(Z[:, 0], Z[:, 1], c=y, alpha=0.7, edgecolors='black');
```
