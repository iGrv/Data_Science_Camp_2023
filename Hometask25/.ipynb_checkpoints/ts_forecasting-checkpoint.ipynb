## Home task: Time Series Forecasting (TS Forecasting)

- Choose any store from the initial dataset
- Check the presence of nans and fill them
- Make a forecast for 30, 180, 270, 365 days ahead
- Perform model evaluation

For TS Forecasting we will use the [Kaggle dataset](https://www.kaggle.com/competitions/store-sales-time-series-forecasting/overview), which contains store sales data from Corporaci√≥n Favorita, a large Ecuadorian-based grocery retailer.

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import os

CWD = os.getcwd()                         # Current working directory
DATA_DIR = os.path.join(CWD, 'data')      # Data directory
IMAGES_DIR = os.path.join(CWD, 'images')  # Images directory
SALES_FILE = os.path.join(DATA_DIR, 'train.csv')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# Read the store sales data
data = pd.read_csv(SALES_FILE, index_col='id', parse_dates=['date'])
data.head()
```

```python
def time_series(sales, store=None):
    '''
    Based on `sales` DataFrame, builds a Series of sales for the specified `store`
    :param sales: DataFrame containing store sales data 
    :type sales: pandas DataFrame
    :param store: store number (column `'store_nbr'` in `sales` DataFrame) which
    to choose sales from, if None, then random chosen, defaults to None
    :type store: int or None, optional
    :return: Series of sales for `store` (column `'sales'` in `sales` DataFrame)
    :rtype: pandas Series
    '''
    if store is None:
        store = np.random.choice(sales['store_nbr'].unique())
    
    # Choose the sales data for the specified store
    # and summarize the sales for each day
    sales = sales[sales['store_nbr'] == store]
    return sales.groupby('date')['sales'].sum()
```

```python
# Investigate the store numbers
data['store_nbr'].unique()
```

So, store numbers are natural numbers from 1 to 54. Let's choose the sales data for store number 30.

```python
STORE = 30

# Get sales data for store 30
ts = time_series(data, store=STORE)
ts.head()
```

```python
# Set zero values to NA
ts[ts == 0] = np.NaN
ts.head()
```

```python
# Convert the TS to daily data
ts = ts.asfreq('D')
print('First 5 values:')
print(ts.head())
print()
print('Last 5 values:')
print(ts.tail())
```

The TS values are indexed in days from January 1st, 2013 to August 15th, 2017. If the TS does not contain values for some days, the converted TS will have NA values for those days. We should check the TS for missing values (NA values).

```python
# Check for missing values
ts.isna().sum()
```

The TS has 32 missing values. To check if days having missing values are holidays, we can load the data about holidays in Ecuador.

```python
HOLIDAYS_FILE = os.path.join(DATA_DIR, 'holidays_events.csv')

# Load the data about holidays
holidays = pd.read_csv(HOLIDAYS_FILE, parse_dates=['date'])
holidays_sales = pd.merge(ts, holidays, how='left', left_on=ts.index, right_on='date')

# Check if days with missing sales values are holidays
holidays_sales[holidays_sales['sales'].isna()]
```

So, there are 9 holidays: 5 days fall on New Years (Sp. *Primer dia del ano*) and 4 days fall on Christmas (Sp. *Navidad*). We does not know why there are no values for other days from July 8th, 2013 to July 30th, 2013.

We will assume that the store was closed for the holidays and had no sales. Thus, we will set these values to zero. For the other days, we will use linear interpolation to fill in the values.

```python
ts_holidays = holidays_sales.loc[holidays_sales['sales'].isna(), ['date', 'description']]

# Determine holiday dates
holiday_dates = ts_holidays[ts_holidays['description'].apply(lambda x: x in ['Primer dia del ano', 'Navidad'])]['date']

# Determine other dates
other_dates = ts_holidays[ts_holidays['description'].apply(lambda x: x not in ['Primer dia del ano', 'Navidad'])]['date']

# Set zero sales for holidays
ts[holiday_dates] = 0

# And use interpolation for other dates
ts = ts.interpolate()
```

### EDA of time series

```python
from statsmodels.tsa.seasonal import seasonal_decompose

# Use classical decomposition to get components of the TS for first 365 days (first year) 
components = seasonal_decompose(ts[:365])

# Plot the components of the TS
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, sharex=True, figsize=(10, 6))
fig.suptitle('Time Series Components')

ax1.set_ylabel('Observed')
sns.lineplot(components.observed, ax=ax1)

ax2.set_ylabel('Trend')
sns.lineplot(components.trend, ax=ax2)

ax3.set_ylabel('Seasonal')
sns.lineplot(components.seasonal, ax=ax3)

ax4.set_ylabel('Residuals')
sns.lineplot(components.resid, ax=ax4);
```

The plots above show us the values of the TS, values of its trend component, values of its seasonal component and residuals. Here we can see that the TS has a seasonal behavior, and the seasonal period is a week, i.e., 7 days.

Now we can draw a box plot to see the distributions of sales for each day of week.

```python
# Add days of week to the TS values
ts_data = pd.DataFrame({'date': ts.index, 'sales': ts.values})
ts_data['day_of_week'] = ts_data['date'].dt.day_name()

# Show a box plot of sales for each day of week
plt.figure(figsize=(9, 6))
sns.boxplot(ts_data, hue='day_of_week', y='sales', gap=0.1)
plt.ylabel('Sales')
plt.legend(title='Day of week', loc='upper right');
```

In the box plot we see the store has the most sales on Tuesday, Saturday, and Sunday.

We also should to find out if the TS is stationary. To do this, we will use the augmented Dickey‚ÄìFuller (ADF) test which null hypothesis states that there exists a unit root in the TS and it is non-stationary. An alternate hypothesis states that there exists no unit root in the TS and it is stationary. To the null hypothesis be rejected and accept the TS be considered as stationary, calculated p-value should be less than a significance level 0.05. Otherwise, the alternate hypothesis will be rejected and the TS will be considered as non-stationary.

```python
from statsmodels.tsa.stattools import adfuller

# Apply the ADF test to find out if the TS is stationary
result = adfuller(ts)
print('Augmented Dickey‚ÄìFuller test:')
print(f'p-value: {result[1]:.4f}')
```

p-value is greater the significance level 0.05, so, we reject the alternate hypothesis: the TS is non-stationary.

### Forecasting

To evaluate the forcast we use three types of forecast errors:
 - Mean absolute error (MAE) - arithmetic average of the absolute errors (differences between actual values $y_i$ and forecast values $\hat{y}_i$)

$$\text{MAE} = \frac{1}{n} \sum_{i = 1}^{n} \left|y_i - \hat{y}_i \right|$$

 - Mean squared error (MSE) - arithmetic average of the squared errors

$$\text{MSE} = \frac{1}{n} \sum_{i = 1}^{n} \left(y_i - \hat{y}_i \right)^2$$

 - Mean absolute percentage error (MAPE) - arithmetic average of the absolute values of errors divided by actual values $y_i$

$$\text{MAPE} = \frac{1}{n} \sum_{i = 1}^{n} \left|\frac{y_i - \hat{y}_i}{y_i} \right|$$

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_absolute_percentage_error

def evaluate(actual, forecast, metrics=None):
    '''
    Perform the forecast evaluation using forecast errors MAE, MSE, and MAPE
    :param actual: truth time series values
    :type actual: pandas Series
    :param forecast: predicted time series values
    :type forecast: pandas Series
    :param metrics: sequence of forecast errors to apply, contains `'mae'`, `'mae'`, or
    `'mape'`, if None, then is `['mae', 'mse', 'mape']`, defaults to None
    :type metrics: iterable or None, optional
    '''
    if metrics is None:
        metrics = ['mae', 'mse', 'mape']

    errors = {
        'mae': mean_absolute_error,
        'mse': mean_squared_error,
        'mape': mean_absolute_percentage_error
    }

    # Calculate an error and print it
    for metric in metrics:
        print(f'{metric.upper()}: {errors[metric](actual, forecast):.4f}')
```

```python
# Split the TS values in training set (all days except last 365 (a year)) and testing set (last 365 days)
ts_train, ts_test = ts[:-365], ts[-365:]

# Construct testing values to get forecast in 30, 180, and 270 days ahead
ts_test_30 = ts[-365:-365 + 30]
ts_test_180 = ts[-365:-365 + 180]
ts_test_270 = ts[-365:-365 + 270]
```

#### ARIMA model

[ARIMA model](https://otexts.com/fpp3/non-seasonal-arima.html) (autoregressive integrated moving average) can be considered as combination of time series differencing with autoregression (AR) and a moving average (MA) model. The full model can be written as

$$(1 - \phi_1 B - ... - \phi_p B^p) (1 - B)^d y_t = c + (1 + \theta_1 B + ... + \theta_q B^q) \epsilon_t $$

where
 - $c$ is constant
 - $\epsilon_t$ is white noise
 - $B y_t = y_{t - 1}$ is backward shift operator, so $(1 - B)^d y_t$ is $d$th-order difference
 - $(1 - \phi_1 B - ... - \phi_p B^p)$ is AR model part
 - $(1 + \theta_1 B + ... + \theta_q B^q) \epsilon_t$ is MA model part

The ARIMA model can often be written as $\text{ARIMA}(p, d, q)$ where
 - $p$ is order of AR part
 - $d$ is differencing order
 - $q$ is order of MA part

The ARIMA model requires the data to be stationary. To achieve the stationarity of the data, differencing is used in the model.

To be able to model seasonal data, ARIMA model includes additional seasonal terms. [Seasonal ARIMA model](https://otexts.com/fpp3/seasonal-arima.html) is written as $\text{ARIMA}(p, d, q)(P, D, Q)_m$ where
 - $P$, $D$,and $Q$ are seasonal parameters similar to those non-seasonal
 - $m$ is seasonal period

The seasonal part of the model consists of terms that are similar to the non-seasonal components of the model, but involve backshifts of the seasonal period. The additional seasonal terms are simply multiplied by the non-seasonal terms.

```python
from pmdarima.arima import auto_arima

# To find out all pararameters of the ARIMA model, use auto_arima function setting m (seasonal period) to 7
model_arima = auto_arima(ts_train, m=7, stepwise=False, n_jobs=-1)
model_arima.summary()
```

So, for given TS the best model is $\text{ARIMA}(1, 1, 2)(1, 0, 1)_7$. Let's build the forecasts and evaluate them.

##### 30 days ahead

```python
# Get point forecasts and forecast intervals (with 90% confidence) in 30 days ahead
pred, interval = model_arima.predict(n_periods=30, return_conf_int=True, alpha=0.1)
forecast = pd.DataFrame({'pred': pred, 'lower': interval[:, 0], 'upper': interval[:, 1]}).reset_index(names='date')

evaluate(ts_test_30, forecast['pred'])
```

Let's visualize the forecast.

```python
plt.figure(figsize=(9, 4))

# Draw the observed values of the TS
ax = sns.lineplot(ts[-365 - 60:-365 + 30], color='g', marker='o', label='Observed values')

# Draw the point forecasts and forecast intervals
sns.lineplot(forecast, x='date', y='pred', color='b', marker='o', label='Forecast', ax=ax)
ax.fill_between(
    forecast['date'], forecast['lower'], forecast['upper'],
    color='b', alpha=0.2, label='Forecast interval (90% confidence)'
)

ax.set_title('ARIMA model forecast (30 days ahead)')
ax.set_xlabel('Date')
ax.set_ylabel('Sales')
ax.legend(loc='lower left')
ax.set_ylim(1000, 9000)

# Save the plot
plt.savefig(os.path.join(IMAGES_DIR, 'arima_forecast.png'))
```

##### 180 days ahead

```python
# Get the point forecasts in 180 days ahead
forecast = model_arima.predict(n_periods=180)

evaluate(ts_test_180, forecast, metrics=['mae', 'mse'])
```

##### 270 days ahead

```python
# Get the point forecasts in 270 days ahead
forecast = model_arima.predict(n_periods=270)

evaluate(ts_test_270, forecast, metrics=['mae', 'mse'])
```

##### 365 days ahead

```python
# Get the point forecasts in 365 days ahead
forecast = model_arima.predict(n_periods=365)

evaluate(ts_test, forecast, metrics=['mae', 'mse'])
```

#### ETS model

ETS models use the [exponential smoothing methods](https://otexts.com/fpp3/ses.html) whose main idea is to calculate the forecast as a a weighted average between the most recent observation nd the previous forecast.

Each [ETS model](https://otexts.com/fpp3/ets.html) has a measurement equation discribing the observed data and some state equations discribing how the unobserved components or states (level, trend, seasonal) change over time. Also model supports one of two types of errors: additive errors or multiplicative errors.

ETS means *Error*, *Trend*, *Seasonal*. The models are labels as $\text{ETS}(error, trend, seasonal)$ where
 - $error = \{\text{A}, \text{M}\}$ is errors component that can be additive $\text{A}$ or multiplicative $\text{M}$
 - $trend = \{\text{N}, \text{A}, \text{A}_d\}$ is trend component (none $\text{N}$, additive $\text{A}$, or additive damped $\text{A}_d$)
 - $seasonal = \{\text{N}, \text{A}, \text{M}\}$ is seasonal component (none $\text{N}$, additive $\text{A}$, or multiplicative $\text{M}$)

For each possible combination of components model has specified equations.

```python
from statsmodels.tsa.exponential_smoothing.ets import ETSModel

# Build the ETS model selecting additive seasonal component (since not all TS values are positive)
# and setting the seasonal period to 7
model_ets = ETSModel(ts_train, seasonal='add', seasonal_periods=7)
model_ets = model_ets.fit()
model_ets.summary()
```

We have built the model $\text{ETS}(\text{A}, \text{N}, \text{A})$  which has additive errors, no trend, and additive seasonal component. It is described by the following equations:

measurement equation $y_t = \ell_{t-1} + s_{t-m} + \epsilon_t$

level equation $\ell_t = \ell_{t-1} + \alpha \epsilon_t$

seasonal equation $s_t = s_{t-m} + \gamma \epsilon_t$

Here
 - $\ell_t$ is level value
 - $s_t$ is seasonal value
 - $m$ is seasonal period
 - $\epsilon_t$ is white noise
 - $\alpha$ is level parameter
 - $\gamma$ is seasonal parameter
 
 Let's build the forecasts and evaluate them.

##### 30 days ahead

```python
# Get point forecasts and forecast intervals (with 90% confidence) in 30 days ahead
predictions = model_ets.get_prediction(start=ts_test_30.index[0], end=ts_test_30.index[-1])
intervals = predictions.pred_int(alpha=0.1).set_axis(['lower', 'upper'], axis=1)
forecast = pd.concat([predictions.predicted_mean.rename('pred'), intervals], axis=1).reset_index(names='date')

evaluate(ts_test_30, forecast['pred'])
```

Let's visualize the forecast.

```python
plt.figure(figsize=(9, 4))

# Draw the observed values of the TS
ax = sns.lineplot(ts[-365 - 60:-365 + 30], color='g', marker='o', label='Observed values')

# Draw the point forecasts and forecast intervals
sns.lineplot(forecast, x='date', y='pred', color='b', marker='o', label='Forecast', ax=ax)
ax.fill_between(
    forecast['date'], forecast['lower'], forecast['upper'],
    color='b', alpha=0.2, label='Forecast interval (90% confidence)'
)

ax.set_title('ETS model forecast (30 days ahead)')
ax.set_xlabel('Date')
ax.set_ylabel('Sales')
ax.legend(loc='lower left')
ax.set_ylim(1000, 9000)

# Save the plot
plt.savefig(os.path.join(IMAGES_DIR, 'ets_forecast.png'))
```

##### 180 days ahead

```python
# Get the point forecasts in 180 days ahead
forecast = model_ets.forecast(steps=180)

evaluate(ts_test_180, forecast, metrics=['mae', 'mse'])
```

##### 270 days ahead

```python
# Get the point forecasts in 270 days ahead
forecast = model_ets.forecast(steps=270)

evaluate(ts_test_270, forecast, metrics=['mae', 'mse'])
```

##### 365 days ahead

```python
# Get the point forecasts in 365 days ahead
forecast = model_ets.forecast(steps=365)

evaluate(ts_test, forecast, metrics=['mae', 'mse'])
```

#### Prophet model

[Prophet](https://facebook.github.io/prophet/) is a library introduced by Facebook for forecasting time series data using an additive model where non-linear trends are fit with yearly, weekly, and daily seasonality, plus holiday effects. It works best with time series that have strong seasonality and several seasons of historical data.

```python
# Convert TS to a DataFrame acceptable for the Prohet model
ts_df = pd.DataFrame({'ds': ts.index, 'y': ts.values})
ts_df.head()
```

```python
# Split the DataFrame in training and testing set
ts_train_df, ts_test_df = ts_df[:-365], ts_df[-365:]

# Construct testing sets to get 30, 180, and 270 days ahead forecasts
ts_test_df_30 = ts_df[-365:-365 + 30]
ts_test_df_180 = ts_df[-365:-365 + 180]
ts_test_df_270 = ts_df[-365:-365 + 270]
```

```python
# Get holiday data to fit it in the Prophet model
holidays_df = holidays[['date', 'description']]
holidays_df = holidays_df.rename(columns={'date': 'ds', 'description': 'holiday'})
holidays_df.head()
```

```python
from prophet import Prophet

# Build the Prophet model fitting in it the holiday data,
# setting up the weekly seasonality and the confidence level of forecast interval to 90%
model_prophet = Prophet(weekly_seasonality=True, interval_width=0.9, holidays=holidays_df)
model_prophet = model_prophet.fit(ts_train_df)
```

##### 30 ahead

```python
# Get point forecasts and forecast intervals (with 90% confidence) in 30 days ahead
forecast = model_prophet.predict(ts_test_df_30[['ds']])

evaluate(ts_test_df_30['y'], forecast['yhat'])
```

Let's visualize the forecast.

```python
plt.figure(figsize=(9, 4))

# Draw the observed values of the TS
ax = sns.lineplot(ts_df[-365 - 60:-365 + 30], x='ds', y='y', color='g', marker='o', label='Observed values')

# Draw the point forecasts and forecast intervals
sns.lineplot(forecast, x='ds', y='yhat', color='b', marker='o', label='Forecast', ax=ax)
ax.fill_between(
    forecast['ds'], forecast['yhat_lower'], forecast['yhat_upper'],
    color='b', alpha=0.2, label='Forecast interval (90% confidence)'
)

ax.set_title('Prophet model forecast (30 days ahead)')
ax.set_xlabel('Date')
ax.set_ylabel('Sales')
ax.legend(loc='lower left')
ax.set_ylim(1000, 9000)

# Save the plot
plt.savefig(os.path.join(IMAGES_DIR, 'prophet_forecast.png'))
```

##### 180 ahead

```python
# Get the point forecasts in 180 days ahead
forecast = model_prophet.predict(ts_test_df_180[['ds']])

evaluate(ts_test_df_180['y'], forecast['yhat'], metrics=['mae', 'mse'])
```

##### 270 ahead

```python
# Get the point forecasts in 270 days ahead
forecast = model_prophet.predict(ts_test_df_270[['ds']])

evaluate(ts_test_df_270['y'], forecast['yhat'], metrics=['mae', 'mse'])
```

##### 365 ahead

```python
# Get the point forecasts in 365 days ahead
forecast = model_prophet.predict(ts_test_df[['ds']])

evaluate(ts_test_df['y'], forecast['yhat'], metrics=['mae', 'mse'])
```
