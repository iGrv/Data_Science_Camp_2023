#### Thyroid Disease dataset

The Thyroid Disease dataset (ann thyriod) is a [dataset from UCI machine learning repository](https://archive.ics.uci.edu/dataset/102/thyroid+disease) which is used for classification task with artificial neural networks (ANNs) training. It has 3772 training instances and 3428 testing instances, and also has 15 categorical and 6 real attributes. The problem is to determine whether a patient referred to the clinic is hypothyroid. Therefore three classes are built: normal (not hypothyroid), hyperfunction and subnormal functioning. For outlier detection, 3772 training instances are used, with only 6 real attributes. The hyperfunction class is treated as outlier class and other two classes are inliers, because hyperfunction is a clear minority class. The dataset is stored in the [annthyroid.mat](data/annthyroid.mat) file.

```python
import os

# Paths to the dataset files
DATA_DIR = 'data'
THYROID_FILE = DATA_DIR + os.path.sep + 'annthyroid.mat'
SATELLITE_FILE = DATA_DIR + os.path.sep + 'satellite.mat'
```

```python
from scipy.io import loadmat

# Load the thyroid dataset
mat_thyroid = loadmat(THYROID_FILE)
thyroid_data = mat_thyroid['X']      # Multi-dim data
thyroid_outliers = mat_thyroid['y']  # Outlier label where 1 = outlier, 0 = not outlier

# Compute the proportion of outliers (in the thyroid dataset it is 7.42%)
thyroid_prop = round(thyroid_outliers.mean(), 4)
print(f'Proportion of outliers in thyroid dataset is about {thyroid_prop} (or {thyroid_prop:.2%})')
```

```python
from sklearn.preprocessing import MinMaxScaler

# Normalize the thyroid dataset data using min-max scaling
scaler = MinMaxScaler()
thyroid_data_scaled = scaler.fit_transform(thyroid_data)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

def evaluate(y_true, y_pred, model_name, labels=None):
    '''
    Builds the confusion matrix, also computes the accuracy,
    precision, recall scores based in truth and predicted labels.

    :param y_true: truth labels
    :type y_true: numpy ndarray of shape `(m,)`
    :param y_pred: predicted labels
    :type y_pred: numpy ndarray of shape `(m,)`
    :param model_name: name of the model which predicted the labels
    :type model_name: str
    :param labels: label names, defaults to None
    :type labels: List[str] or None, optional
    '''
    # Build the confusion matrix
    conf_mat = confusion_matrix(y_true, y_pred)
    tn, fp, fn, tp = conf_mat.ravel()

    # Compute scores based on true positive (TP), true negative (TN),
    # false positive (FP) and false negative (FN) values
    accuracy = (tn + tp) / (tn + fp + fn + tp)
    precision = tp / (tp + fp)
    recall = tp / (tp + fn)

    # Print the scores and show the confusion matrix
    print(model_name)
    print(f'- accuracy: {accuracy:.4f}')
    print(f'- precision: {precision:.4f}')
    print(f'- recall: {recall:.4f}')

    disp = ConfusionMatrixDisplay(conf_mat, display_labels=labels).plot()
    disp.ax_.set_title(f'Confusion matrix ({model_name})')
```

#### Elliptic Envelope (thyroid dataset)

```python
from sklearn.covariance import EllipticEnvelope

# Use the Elliptic Envelope model to detect the outliers
model = EllipticEnvelope(contamination=thyroid_prop, random_state=0)
preds = model.fit_predict(thyroid_data_scaled)
outlier_preds = (preds == -1).astype('int')

# Perform the model evaluation
evaluate(thyroid_outliers, outlier_preds, 'Elliptic Envelope on thyroid dataset', labels=['not outlier', 'outlier']);
```

#### Isolation Forest (thyroid dataset)

```python
from sklearn.ensemble import IsolationForest

# Use the Isolation Forest model to detect the outliers
model = IsolationForest(n_estimators=200, contamination=thyroid_prop, random_state=0)
preds = model.fit_predict(thyroid_data_scaled)
outlier_preds = (preds == -1).astype('int')

# Perform the model evaluation
evaluate(thyroid_outliers, outlier_preds, 'Isolation Forest on thyroid dataset', labels=['not outlier', 'outlier']);
```

#### Local Outlier Factor (thyroid dataset)

```python
from sklearn.neighbors import LocalOutlierFactor

# Use the Local Outlier Factor model to detect the outliers
model = LocalOutlierFactor(n_neighbors=100, contamination=thyroid_prop)
preds = model.fit_predict(thyroid_data_scaled)
outlier_preds = (preds == -1).astype('int')

# Perform the model evaluation
evaluate(thyroid_outliers, outlier_preds, 'Local Outlier Factor on thyroid dataset', labels=['not outlier', 'outlier']);
```

#### One-Class SVM (thyroid dataset)

```python
from sklearn.svm import OneClassSVM

# Use the One-Class SVM model to detect the outliers
model = OneClassSVM(kernel='rbf', gamma=0.5)
preds = model.fit_predict(thyroid_data_scaled)
outlier_preds = (preds == -1).astype('int')

# Perform the model evaluation
evaluate(thyroid_outliers, outlier_preds, 'One-Class SVM on thyroid dataset', labels=['not outlier', 'outlier']);
```

#### DBSCAN (thyroid dataset)

```python
from sklearn.cluster import DBSCAN

# Use the DBSCAN model to detect the outliers
model = DBSCAN(eps=0.05)
preds = model.fit_predict(thyroid_data_scaled)
outlier_preds = (preds == -1).astype('int')

# Perform the model evaluation
evaluate(thyroid_outliers, outlier_preds, 'DBSCAN on thyroid dataset', labels=['not outlier', 'outlier']);
```

#### Satellite dataset

The Statlog (Landsat Satellite) dataset is a [dataset from UCI machine learning repository](https://archive.ics.uci.edu/dataset/146/statlog+landsat+satellite) which is used for multi-class classification. Here, the training and test data are combined.

The dataset consists of the multi-spectral values of pixels in 3x3 neighbourhoods in a satellite image, and the classification associated with the central pixel in each neighbourhood. The aim is to predict this classification, given the multi-spectral values. In the sample database, the class of a pixel is coded as a number.

The dataset is a (tiny) sub-area of a scene, consisting of 82 x 100 pixels. Each line of data corresponds to a 3x3 square neighbourhood of pixels completely contained within the 82x100 sub-area. Each line contains the pixel values in the four spectral bands (converted to ASCII) of each of the 9 pixels in the 3x3 neighbourhood and a number indicating the classification label of the central pixel. The attributes are numerical, in the range 0 to 255. The number is a code for the following classes:

1. red soil
2. cotton crop
3. grey soil
4. damp grey soil
5. soil with vegetation stubble
6. mixture class (all types present)
7. very damp grey soil

For outlier detection, the smallest three classes, i.e. 2, 4, 5 are combined to form the outliers class, while all the other classes are combined to form an inlier class. The dataset is stored in the [satellite.mat](data/satellite.mat) file.

```python
# Load the satellite dataset
mat_satellite = loadmat(SATELLITE_FILE)
satellite_data = mat_satellite['X']      # Multi-dim data
satellite_outliers = mat_satellite['y']  # Outlier label where 1 = outlier, 0 = not outlier

# Compute the proportion of outliers (in the satellite dataset it is 32%)
satellite_prop = round(satellite_outliers.mean(), 2)
print(f'Proportion of outliers in satellite dataset is about {satellite_prop} (or {satellite_prop:.0%})')
```

```python
# Normalize the satellite dataset data so that the data is in the range 0 to 1
satellite_data_scaled = satellite_data / 255
```

```python
# Use the Elliptic Envelope model to detect the outliers
model = EllipticEnvelope(contamination=satellite_prop, random_state=0)
preds = model.fit_predict(satellite_data_scaled)
outlier_preds = (preds == -1).astype('int')

# Perform the model evaluation
evaluate(satellite_outliers, outlier_preds, 'Elliptic Envelope on satellite dataset', labels=['not outlier', 'outlier']);
```

#### Isolation Forest (satellite dataset)

```python
# Use the Isolation Forest model to detect the outliers
model = IsolationForest(n_estimators=500, contamination=satellite_prop, random_state=0)
preds = model.fit_predict(satellite_data_scaled)
outlier_preds = (preds == -1).astype('int')

# Perform the model evaluation
evaluate(satellite_outliers, outlier_preds, 'Isolation Forest on satellite dataset', labels=['not outlier', 'outlier']);
```

#### Local Outlier Factor (satellite dataset)

```python
# Use the Local Outlier Factor model to detect the outliers
model = LocalOutlierFactor(n_neighbors=1000, contamination=satellite_prop)
preds = model.fit_predict(satellite_data_scaled)
outlier_preds = (preds == -1).astype('int')

# Perform the model evaluation
evaluate(satellite_outliers, outlier_preds, 'Local Outlier Factor on satellite dataset', labels=['not outlier', 'outlier']);
```
