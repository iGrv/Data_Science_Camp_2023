# Home task: Clustering Algorithms

Apply various clustering algorithms 

- Agglomerative clustering using `complete` linkage
- Agglomerative clustering using `single` linkage
- DBSCAN
- Use Elbow method to determine the best k for Kmeans

## Data

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs
%matplotlib inline
```

```python
X, y = make_blobs(n_samples=500, n_features=2, centers=5, cluster_std=0.6, random_state=0)
colors = np.array([
    plt.cm.Spectral(val)
    for val in np.linspace(0, 1, len(set(y)))
])

plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=colors[y], s=20);
```

## Agglomerative clustering using `complete` linkage

```python
from sklearn.cluster import AgglomerativeClustering

# Perform agglomerative clustering to get 5 clusters using complete linkage
model = AgglomerativeClustering(n_clusters=5, linkage='complete')
aggl_complete_pred = model.fit_predict(X)

plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=colors[aggl_complete_pred], s=20);
```

## Agglomerative clustering using `single` linkage

```python
# Perform agglomerative clustering to get 5 clusters using single linkage
model = AgglomerativeClustering(n_clusters=5, linkage='single')
aggl_single_pred = model.fit_predict(X)

plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=colors[aggl_single_pred], s=20);
```

Let's use aggregate clustering using `ward` linkage to get better clustering results.

```python
# Perform agglomerative clustering to get 5 clusters using ward linkage
model = AgglomerativeClustering(n_clusters=5, linkage='ward')
aggl_ward_pred = model.fit_predict(X)

plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=colors[aggl_ward_pred], s=20);
```

## DBSCAN

```python
from sklearn.cluster import DBSCAN

# Perform DBSCAN clustering with maximum distance between two samples of
# same cluster of 0.6 and minimum number of samples in a cluster of 10
model = DBSCAN(eps=0.6, min_samples=10)
dbscan_pred = model.fit_predict(X)

# Add grey color for outliers since the outlier predictions are equal to -1 (refferes to the last element of the array)
colors = np.vstack([colors, plt.cm.Greys(0.7)])

plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=colors[dbscan_pred], s=20);
```

Let's compare the clustering algorithm by building scatter plots for each of them.

```python
def scatter_plots(x, y, n_rows, n_cols, titles, colors):
    for i in range(n_rows * n_cols):
        plt.subplot(n_rows, n_cols, i + 1)
        plt.title(titles[i])
        plt.scatter(x, y, c=colors[i], s=15)
```

```python
# Titles for each algorithm
titles = [
    'Agglomerative clustering (complete linkage)',
    'Agglomerative clustering (single linkage)',
    'Agglomerative clustering (ward linkage)',
    'DBSCAN'
]

# Colors to represent clusters for each algorithm
plot_colors = [
    colors[pred]
    for pred in [aggl_complete_pred, aggl_single_pred, aggl_ward_pred, dbscan_pred]
]

plt.figure(figsize=(11, 9))
scatter_plots(X[:, 0], X[:, 1], 2, 2, titles, plot_colors);
```

## KMeans with Elbow method

```python
from sklearn.cluster import k_means

# Consider k in range from 1 to 7 inclusive
k_values = list(range(1, 8))

# Compute final cost values (final values of inertia criterion) in KMeans clustering for different k
costs = []
for k in k_values:
    _, _, cost = k_means(X, k, random_state=0)
    costs.append(cost)

# Plot final cost values for different k to select best value of k
plt.title('Finding the best k for KMeans algorithm')
plt.xlabel('k')
plt.ylabel('Final cost')
plt.plot(k_values, costs, color='red', marker='o');
```

In the plot we can see that the final cost value does not decrease much for k = 3 and k = 4, so the best number of clusters is k = 3, k = 4. After k = 5, final cost value does not decrease at all.
