##  Brest cancer  dataset
</font>



```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split

# https://scikit-learn.org/stable/datasets/toy_dataset.html#breast-cancer-dataset
cancer = load_breast_cancer()
X, y = cancer['data'], cancer['target']
features, labels = cancer['feature_names'], cancer['target_names']
print('features:', features)
print('labels:', labels)

# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

<font color = green>

##  DecisionTree 

</font>



```python
# YOUR_CODE.  Create the classifier and train it 
# START_CODE 
import graphviz
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
from IPython.display import Image

# Create decision tree classifier which uses entropy and information gain
# to measure the tree splits, and limits the depth of tree to 3
clf = DecisionTreeClassifier(
    criterion='entropy',
    max_depth=3,
    random_state=0
).fit(X_train, y_train)
# END_CODE 

print('train accuracy = {:.3%}'.format(clf.score(X_train, y_train)))
print('test accuracy = {:.3%}'.format(clf.score(X_test, y_test)))

graph_viz = export_graphviz(clf, feature_names=features, class_names=labels, filled=True, rounded=True)
graph = graphviz.Source(graph_viz)
graph_file = graph.render(format='png', cleanup=True)  # cleanup (bool)  Delete the source file after rendering.
Image(filename=graph_file)
```

<font color = green>

##  RandomForest

</font>



```python
# YOUR_CODE.  Create the classifier and train it 
# START_CODE 
from sklearn.ensemble import RandomForestClassifier

# Create random forest classifier which
# builds 250 trees with depth limit to 4
clf = RandomForestClassifier(
    n_estimators=250,
    max_depth=4,
    random_state=0
).fit(X_train, y_train)
# END_CODE 

print('train accuracy = {:.3%}'.format(clf.score(X_train, y_train)))
print('test accuracy = {:.3%}'.format(clf.score(X_test, y_test)))
```

<font color = green>

##  Gradient Boosting Decision Trees (GBDT)

</font>



```python
# YOUR_CODE.  Create the classifier and train it 
# START_CODE 
from sklearn.ensemble import GradientBoostingClassifier

# Create gradient boosting classifier which uses
# a learning rate of 0.2 and limits the depth of trees to 2
clf = GradientBoostingClassifier(
    learning_rate=0.2,
    max_depth=2,
    random_state=0
).fit(X_train, y_train)
# END_CODE 

print('train accuracy = {:.3%}'.format(clf.score(X_train, y_train)))
print('test accuracy = {:.3%}'.format(clf.score(X_test, y_test)))
```

<font color = green>

##  XGBoost

</font>



```python
# YOUR_CODE.  Create the classifier and train it 
# START_CODE 
from xgboost import XGBClassifier

# Create XGBoost classifier which limits the depth of trees to 3
# and uses the L2 regularization with a coefficient of 10
clf = XGBClassifier(
    max_depth=3,
    reg_lambda=10,
    random_state=0
).fit(X_train, y_train)
# END_CODE 

print('train accuracy = {:.3%}'.format(clf.score(X_train, y_train)))
print('test accuracy = {:.3%}'.format(clf.score(X_test, y_test)))
```
